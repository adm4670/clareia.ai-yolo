```python
pdf_path = "./provas/2013/Caderno1_Azul_Sab.pdf"
```

```python
pdf_document = fitz.open(pdf_path)
```

```python
pagina = pdf_document[2]
```

```python
blocos = pagina.get_text("blocks")
```

```python
padrao_questao = r"QUESTÃO\s*0*(\d{1,2})"
```

```python
pagina_dict = pagina.get_text("dict")

imprimir = False

for bloco in pagina_dict["blocks"]:
    if bloco["type"] == 0:  # bloco de texto
        # Extrair todo o texto do bloco mantendo quebras de linha
        linhas_texto = []
        for linha in bloco["lines"]:
            linha_texto = "".join([span["text"] for span in linha["spans"]])
            linhas_texto.append(linha_texto)
        conteudo = "\n".join(linhas_texto)  # mantém a formatação original
        
        # Verifica se é a questão
        match = re.search(padrao_questao, conteudo, flags=re.IGNORECASE)
        if match:
            numero_questao = int(match.group(1))
            imprimir = True
            print(f"----------- QUESTÃO {numero_questao} -----------")
            print(f"### {conteudo}\n")
            continue
        
        # Imprime blocos subsequentes
        if imprimir:
            print(conteudo)

    if bloco["type"] == 1:
        print("TEM IMAGEM AQUI")
```
