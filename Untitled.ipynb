```python
question_to_json_agent = Agent(
    name="ENEMQuestionToJSON",
    role="Estruturador de dados educacionais",
    goal=(
        "Converter questões do ENEM extraídas de PDF em um JSON estruturado, "
        "padronizado e pronto para uso na plataforma clareia.ai."
    ),
    backstory=(
        "Especialista em avaliação educacional, editoração técnica e modelagem "
        "de dados para plataformas educacionais baseadas em diagnóstico e personalização."
    ),
    model="gpt-4o-mini",
    general_instructions=(
        "TAREFA ÚNICA:\n"
        "- Receber o texto bruto de uma questão do ENEM extraída de PDF\n"
        "- Converter para JSON estruturado conforme o formato abaixo\n\n"

        "REGRAS OBRIGATÓRIAS (NÃO NEGOCIÁVEIS):\n"
        "- NÃO resolva a questão\n"
        "- NÃO indique alternativa correta\n"
        "- NÃO explique o conteúdo\n"
        "- NÃO altere palavras, números ou frases do enunciado\n"
        "- NÃO reescreva alternativas\n"
        "- NÃO crie informações que não estejam explícitas\n"
        "- NÃO inclua comentários fora do JSON\n\n"

        "VOCÊ PODE:\n"
        "- Organizar o texto em campos estruturados\n"
        "- Separar textos base, enunciado e alternativas\n"
        "- Remover cabeçalhos, rodapés e textos editoriais do PDF\n"
        "- Inferir APENAS metadados estruturais e pedagógicos leves\n\n"

        "TEXTOS QUE DEVEM SER REMOVIDOS:\n"
        "- Cabeçalhos (ex: ENEM 2021 – Caderno Azul)\n"
        "- Rodapés e numeração de página\n"
        "- Créditos editoriais (Fonte, Inep, Adaptado de...)\n"
        "- Marcas de diagramação do PDF\n\n"

        "FORMATO JSON OBRIGATÓRIO (RETORNE EXATAMENTE ISSO):\n"
        "{\n"
        '  "id": "string | identificador único da questão",\n'
        '  "prova": "ENEM",\n'
        '  "ano": number,\n'
        '  "caderno": "string | se identificado, senão null",\n'
        '  "area": "string | área do conhecimento",\n'
        '  "disciplina": "string | se identificável, senão null",\n'
        '  "enunciado": "string | texto principal da questão",\n'
        '  "textos_base": [\n'
        '    "string | texto de apoio 1",\n'
        '    "string | texto de apoio 2"\n'
        "  ],\n"
        '  "imagem": {\n'
        '    "existe": boolean,\n'
        '    "descricao": "string | se houver, senão null",\n'
        '    "referencia": "string | se houver, senão null"\n'
        "  },\n"
        '  "alternativas": [\n'
        '    { "letra": "A", "texto": "string" },\n'
        '    { "letra": "B", "texto": "string" },\n'
        '    { "letra": "C", "texto": "string" },\n'
        '    { "letra": "D", "texto": "string" },\n'
        '    { "letra": "E", "texto": "string" }\n'
        "  ],\n"
        '  "metadados_pedagogicos": {\n'
        '    "tipo_questao": "string | ex: interpretação, cálculo, análise gráfica",\n'
        '    "eixo_tematico": "string | tema geral",\n'
        '    "nivel_leitura": "baixo | medio | alto",\n'
        '    "habilidade_enem": "string | se explícita, senão null",\n'
        '    "competencia_enem": "string | se explícita, senão null"\n'
        "  },\n"
        '  "origem": {\n'
        '    "fonte": "PDF ENEM",\n'
        '    "observacoes": "string | ex: textos editoriais removidos"\n'
        "  }\n"
        "}\n\n"

        "IMPORTANTE:\n"
        "- Se alguma informação não estiver clara no texto, use null\n"
        "- Retorne SOMENTE o JSON válido\n"
        "- Não utilize Markdown\n"
        "- Não inclua qualquer explicação\n"
    ),
)
```

```python
async def formatar_questao_markdown(numero, texto):
    conteudo_bruto = "\n".join(texto)

    prompt = f"""
    A seguir está o texto bruto de uma questão do ENEM extraída de PDF.
    
    QUESTÃO {numero}
    
    {conteudo_bruto}
    """

    resposta = await markdown_formatter_agent.run(prompt)

    # print("\n" + "=" * 60)
    # print(f"QUESTÃO {numero} — MARKDOWN FINAL")
    # print("=" * 60)
    print(resposta)
    print("\n")
```

```python
import fitz  # PyMuPDF
import re

pdf_path = "./provas/2013/Caderno5_Amarelo_Dom.pdf"

pdf_document = fitz.open(pdf_path)

numero_da_pagina_no_pdf = 8
pagina = pdf_document[numero_da_pagina_no_pdf - 1]

padrao_questao = r"QUESTÃO\s*0*(\d{1,2})"

pagina_dict = pagina.get_text("dict")

questao_atual = None
texto_questao = []

def imprimir_questao(numero, texto):
    print("\n" + "=" * 50)
    print(f"QUESTÃO {numero}")
    print("=" * 50)
    print("\n".join(texto))
    print("\n")

for bloco in pagina_dict["blocks"]:
    if bloco["type"] == 0:  # bloco de texto
        linhas_texto = []
        for linha in bloco["lines"]:
            linha_texto = "".join(span["text"] for span in linha["spans"])
            linhas_texto.append(linha_texto)

        conteudo = "\n".join(linhas_texto).strip()

        if not conteudo:
            continue

        match = re.search(padrao_questao, conteudo, flags=re.IGNORECASE)

        if match:
            # Se já estava em uma questão, finaliza e imprime
            if questao_atual is not None:
                await formatar_questao_markdown(questao_atual, texto_questao)

            # Inicia nova questão
            questao_atual = int(match.group(1))
            texto_questao = [conteudo]
        else:
            # Continua acumulando texto da questão atual
            if questao_atual is not None:
                texto_questao.append(conteudo)

    elif bloco["type"] == 1:
        if questao_atual is not None:
            texto_questao.append("[IMAGEM DETECTADA]")

# Imprime a última questão ao final da página
if questao_atual is not None:
    await formatar_questao_markdown(questao_atual, texto_questao)
```
